```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'nonlinear.csv')
plt.scatter(df['x'],df['y'])
```

```python
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# X와 y 데이터를 로드합니다.
# X : feature, y : target
X = df['x'].to_numpy()
y = df['y'].to_numpy()
print(X.shape, y.shape)

# X 데이터를 2차원 배열로 변환합니다.
X = X.reshape(-1, 1)

# 다항식의 차수(degree)가 3인 다항식 피처를 생성하기 위한 PolynomialFeatures 객체를 생성합니다.
poly = PolynomialFeatures(degree=3)

# X 데이터를 다항식 피처로 변환합니다.
X_poly = poly.fit_transform(X)

# 선형 회귀 모델을 생성합니다.
model = LinearRegression()

# 예측을 위한 도메인 데이터를 생성합니다.
domain = np.linspace(0, 1, 100).reshape(-1, 1)

# 생성한 다항식 피처와 y 값을 사용하여 선형 회귀 모델을 학습합니다.
model.fit(X_poly, y)

# 예측을 위해 도메인 데이터를 다항식 피처로 변환합니다.
domain_3 = poly.fit_transform(domain)

# 도메인 데이터에 대한 예측값을 계산합니다.
pred = model.predict(domain_3)

# 학습 데이터와 예측 데이터를 scatter plot으로 시각화합니다.
plt.scatter(df['x'], df['y'])
plt.scatter(domain, pred, color='r')
```
